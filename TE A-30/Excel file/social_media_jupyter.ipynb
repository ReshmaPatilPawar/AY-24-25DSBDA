```python
import matplotlib.pyplot as plt

df["Sentiment"].value_counts().plot(kind="bar", color=["green", "red", "gray"])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Type")
plt.ylabel("Count")
plt.show()
```

```python
df["Platform"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["blue", "purple", "orange"])
plt.title("Platform Distribution")
plt.show()
```

```python
from collections import Counter

all_hashtags = " ".join(df["Hashtags"]).split()  # Combine all hashtags into one list
top_hashtags = Counter(all_hashtags).most_common(10)  # Get top 10 hashtags
hashtags, counts = zip(*top_hashtags)

plt.barh(hashtags, counts, color="cyan")
plt.xlabel("Count")
plt.ylabel("Hashtags")
plt.title("Top 10 Hashtags")
plt.show()
```
